```python
import numpy as np
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': rng.randint(0, 10, 6)},
                   columns = ['key', 'data1', 'data2'])
df
```

```python
df.groupby('key').aggregate([np.median,max]) #aggregate to perform multiple operations on all the columns
```

```python
df.groupby('key').aggregate({'data1':[np.median,max],'data2':np.mean})
```

```python
##transformation
df.groupby('key').transform(lambda x : x - x.mean())
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.groupby('sex')['survived'].mean()
```

```python
titanic.groupby(['sex','class'])['survived'].mean()
```

```python
titanic.groupby(['sex','class']).aggregate({'survived':np.mean})
```

```python
titanic.pivot_table('survived',index='sex',columns='class')
```

```python
titanic['age'].hist(bins=50)
```

```python
titanic.pivot_table('survived', index = ['sex', 'age'], columns = 'class')
```

```python
titanic.pivot_table(index='sex', columns='class',aggfunc={'survived':np.sum})
```
